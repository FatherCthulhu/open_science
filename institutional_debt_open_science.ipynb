```stata
clear all

qui {
    forvalues year = 2021/2021 {

        * 1. Download zip
        copy "https://nces.ed.gov/ipeds/datacenter/data/EF2021A_Data_Stata.zip" ///
            "EF2021A_Data_Stata.zip", replace

        * 2. Unzip (creates EF2021A_Data_Stata.csv)
        unzipfile "EF2021A_Data_Stata.zip", replace

        * 3. Import the CSV that actually exists
        import delimited "EF2021A_Data_Stata.csv", clear varnames(1)

        * 4. Standardize variable names to lowercase
        rename *, lower

        * 5. Filter to level 2 (EF, not EFFY)
        * Check once with: describe ef*
        keep if efalev == 2

        * 6. Keep race/ethnicity totals
        keep unitid efytotlt efyaiant efyasiat efybkaat efyhispt ///
             efynhpit efywhitt efy2mort efyunknt efynralt

        * 7. Rename to ttl*
        foreach var in efytotlt efyaiant efyasiat efybkaat efyhispt ///
                       efynhpit efywhitt efy2mort efyunknt efynralt {
            rename `var' ttl`var'
        }

        * 8. Add year and save
        gen year = `year'
        save ttlef`year', replace
    }
}
```
